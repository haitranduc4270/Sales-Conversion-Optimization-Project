```python
import pandas as pd

file_path = 'data/CitiesExt.csv'
data = pd.read_csv(file_path)

data.head(), data.info()
```

1. Bar chart for the ten countries with the highest population

```python
import matplotlib.pyplot as plt

country_population = data.groupby('country')['population'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
country_population.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries by Population', fontsize=14)
plt.ylabel('Total Population (Millions)', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

2. Pie chart for cities west and east of the Prime Meridian

```python
west_cities = data[data['longitude'] < 0].shape[0]
east_cities = data[data['longitude'] >= 0].shape[0]

plt.figure(figsize=(8, 8))
plt.pie(
    [west_cities, east_cities],
    labels=['West', 'East'],
    autopct='%1.1f%%',
    colors=['#ff9999', '#66b3ff'],
    startangle=140,
    explode=(0.1, 0)
)
plt.title('Proportion of Cities West and East of the Prime Meridian', fontsize=14)
plt.show()
```

3. Scatter plot for latitude vs. population

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['latitude'], data['population'], alpha=0.6, c='green', edgecolor='k')
plt.title('Latitude vs. Population', fontsize=14)
plt.xlabel('Latitude', fontsize=12)
plt.ylabel('Population (Millions)', fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```

The scatterplot doesn't seem to show a clear or consistent relationship between the latitude of cities in a country and the population of those cities. The populations are distributed across all latitudes, with no obvious trend or correlation.

==> This could indicate that population size is not directly influenced by latitude alone and might depend on other factors
